```python
import pandas as pd
```

```python
ratings=[['John', 5,5,2,1], ['Mary',4,5,3,2], ['Bob', 4,4,4,3], ['Lisa',2,2,4,5], ['Lee',1,2,3,4],['Harry',2,1,5,5]]
```

```python
ratings
```

```python
type(ratings)
```

```python
ratings[0]
```

```python
titles=['User','Jaws','Star Wars','Exorcist','Omen']
```

```python
titles
```

```python
movies=pd.DataFrame(ratings, columns=titles) #create a dataframe
```

```python
movies
```

```python
from sklearn import cluster
```

```python
data=movies.drop('User', axis=1) #Drop column user from DF
```

```python
data
```

```python
#initialization of the Kmeans algorithm.
#n_clusters: number of clusters that the algorithm should try to find.
#max_iter: maximum number of iterations

k_means=cluster.KMeans(n_clusters=2, max_iter=100, random_state=1)
```

```python
k_means
```

```python
k_means.fit(data) #Train the model
```

```python
labels=k_means.labels_
```

```python
labels
```

```python
movies
```

```python
outcome=pd.DataFrame(labels, index=movies.User, columns=['Cluster ID'])
```

```python
outcome
```

```python
k_means.cluster_centers_
```
